```python
# Selecione as colunas: loyalty_number, year, month, flights_booked,
#  total_flights, distance e points_accumulated da tabela
#  “flight_activity”.
consulta_atividade = """
SELECT
fa.loyalty_number,
fa.year,
fa.month,
fa.flights_booked,
fa.total_flights,
fa.distance,
fa.points_accumulated
FROM
flight_activity fa
"""

# salvando a consulta dentro da variável df_atividade
df_atividade = pd.read_sql_query( consulta_atividade, conn )

# visualiza os dados
df_atividade.head()
```

```python
#Selecione as mesmas colunas, porém, recupere somente as linhas cuja coluna
#distance é maior do que 2000, da tabela “flight_activity”
consulta_atividade = """
SELECT
fa.loyalty_number,
fa.year,
fa.month,
fa.flights_booked,
fa.total_flights,
fa.distance,
fa.points_accumulated
FROM
flight_activity fa
WHERE
fa.distance > 2000
"""

# salvando a consulta dentro da variável df_atividade
df_atividade = pd.read_sql_query( consulta_atividade, conn )

# visualiza os dados
df_atividade.head()
```

```python
#Selecione as mesmas colunas, porém, recupere somente as linhas cuja coluna
# distance é maior do que 2000 e a coluna month é igual a 9, da tabela
# “flight_activity”

consulta_atividade = """
SELECT
fa.loyalty_number,
fa.year,
fa.month,
fa.flights_booked,
fa.total_flights,
fa.distance,
fa.points_accumulated
FROM
flight_activity fa
WHERE
fa.distance > 2000 AND fa.month = 9
"""

# salvando a consulta dentro da variável df_atividade
df_atividade = pd.read_sql_query( consulta_atividade, conn )

# visualiza os dados
df_atividade.head()
```

```python
# Selecione as mesmas colunas, porém, recupere somente as linhas cuja coluna
# distance é maior do que 2000 ou a coluna points_accumulated é menor que 100,
# da tabela “flight_activity”.

consulta_atividade = """
SELECT
fa.loyalty_number,
fa.year,
fa.month,
fa.flights_booked,
fa.total_flights,
fa.distance,
fa.points_accumulated
FROM
flight_activity fa
WHERE
fa.distance > 2000 OR fa.points_accumulated < 100
"""

# salvando a consulta dentro da variável df_atividade
df_atividade = pd.read_sql_query( consulta_atividade, conn )

# visualiza os dados
df_atividade.head()
```

```python
#Selecione as colunas: loyalty_number, month, year, distance, flights_booked e
#total_flights da tabela “flight_activity”, e recupere somente as linhas cuja
# coluna “flights_booked” é maior do que 10 e a coluna “flights_booked” é
# menor do que 12, da tabela “flight_activity

consulta_atividade = """
SELECT
fa.loyalty_number,
fa.month,
fa.year,
fa.distance,
fa.flights_booked,
fa.total_flights
FROM
flight_activity fa
WHERE
fa.flights_booked > 10 AND fa.flights_booked < 12
"""

# salvando a consulta dentro da variável df_atividade
df_atividade = pd.read_sql_query( consulta_atividade, conn )

# visualiza os dados
df_atividade.head()
```
